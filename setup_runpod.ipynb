# Set Up the Repository on RunPod

Run this notebook once from the RunPod Jupyter environment. It clones the repository into `/workspace/transformers`, or fast-forwards an existing clone to the latest `main` branch.

After the cell completes, use the Jupyter file browser to open `/workspace/transformers/runpod_models.ipynb`.

```python
from pathlib import Path
import subprocess

REPO_URL = "https://github.com/samhatescoding/transformers.git"
REPO_BRANCH = "main"
REPO_ROOT = Path("/workspace/transformers")

if REPO_ROOT.exists():
    if not (REPO_ROOT / ".git").is_dir():
        raise RuntimeError(f"{REPO_ROOT} exists but is not a Git repository.")
    subprocess.run(["git", "-C", str(REPO_ROOT), "fetch", "origin", REPO_BRANCH], check=True)
    subprocess.run(
        ["git", "-C", str(REPO_ROOT), "merge", "--ff-only", f"origin/{REPO_BRANCH}"],
        check=True,
    )
else:
    REPO_ROOT.parent.mkdir(parents=True, exist_ok=True)
    subprocess.run(
        ["git", "clone", "--branch", REPO_BRANCH, "--single-branch", REPO_URL, str(REPO_ROOT)],
        check=True,
    )

notebook_path = REPO_ROOT / "runpod_models.ipynb"
if not notebook_path.exists():
    raise FileNotFoundError(f"Expected notebook was not found: {notebook_path}")

print(f"Repository ready: {REPO_ROOT}")
print(f"Open this notebook next: {notebook_path}")
```
